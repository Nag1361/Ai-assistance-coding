<h1>Task 1: Classes – Student Class
Explanation

In this task, a Student class is created to store and display student information.
The class contains:

Attributes: name, roll number, and branch.

Constructor (__init__): initializes the object values.

Method (display_details): prints student details<h1>

```python
class Student:
    def __init__(self, name, roll_number, branch):
        self.name = name
        self.roll_number = roll_number
        self.branch = branch

    def display_details(self):
        print("Student Name:", self.name)
        print("Roll Number:", self.roll_number)
        print("Branch:", self.branch)


# Object creation
s1 = Student("Nagaraju", 101, "CSE")
s1.display_details()
```

<h2>Analysis

The AI-generated code is correct and easy to understand.
It follows proper object-oriented principles by using a constructor and instance variables.
The method names are meaningful and improve readability.<h2>

<h2>Task 2: Loops – Multiples of a Number
Explanation

This task displays the first 10 multiples of a given number using loop structures.

Using for Loop<h2>

```python
def print_multiples(num):
    for i in range(1, 11):
        print(num * i)

print_multiples(5)
```

```python
def print_multiples_while(num):
    i = 1
    while i <= 10:
        print(num * i)
        i += 1

print_multiples_while(5)
```

<h2>Analysis

The for loop is more suitable when the number of iterations is known in advance.
The while loop requires manual control of the loop variable.
Both approaches generate correct output, but the for loop is simpler and more readable.

Task 3: Conditional Statements – Age Classification
Explanation

This task classifies a person into different age groups using conditional statements.<h2>

<h1>Python Code<h1>

```python
def classify_age(age):
    if age < 13:
        return "Child"
    elif age < 20:
        return "Teenager"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"
```

<h1>Analysis

The conditions are checked sequentially from top to bottom.
Once a condition is satisfied, the corresponding age group is returned.
The AI-generated logic is clear, correct, and well-structured.

Task 4: For and While Loops – Sum of First n Numbers
Explanation

This task calculates the sum of the first n natural numbers.

<h1>
<h1>Using for Loop<h1>

```python
def sum_to_n(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total
```

Using while Loop

```python
def sum_to_n_while(n):
    total = 0
    i = 1
    while i <= n:
        total += i
        i += 1
    return total
```

Using Mathematical Formula

```python
def sum_to_n_formula(n):
    return n * (n + 1) // 2
```

<h1>Analysis

The loop-based approaches calculate the sum by iterating through each number.
The mathematical formula provides the most efficient solution with constant time complexity.
AI correctly suggests both iterative and optimized solutions.

Task 5: Classes – Bank Account Class
Explanation

In this task, a BankAccount class is created to simulate basic banking operations.<h2>

<h2>Python Code<h2>

```python
class BankAccount:
    def __init__(self, account_holder, balance=0):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient balance")

    def check_balance(self):
        print("Current Balance:", self.balance)
```

<h3>
